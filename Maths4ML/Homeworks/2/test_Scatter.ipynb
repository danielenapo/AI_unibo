```python
import numpy as np
import matplotlib.pyplot as plt

#create two cloud of 100 points each in 2D, divided by colors

x1=np.random.normal(0,1,(2,100)) 
x2=np.random.normal(3,0.1,(2,100))

#labels
y1=np.zeros((100,)) #corresponding to x1
y2=np.ones((100,)) #corresponding to x2

#visualization, colouring by class
#join X1 and X2
X=np.concatenate((x1,x2),axis=1) #concatenate BY COLUMN (it's a matrix)
Y=np.concatenate((y1,y2)) #just concatenate (vector)

print(X.shape)
print(Y.shape)

plt.scatter(X[0,:], X[1,:], c=Y) #c means color -> we color the data based on the label
plt.show()
```
